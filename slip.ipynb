```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Create a DataFrame from the given data
data = {
    "Country": ["France", "Spain", "Germany", "Spain", "Germany", "France", "Spain", "France", "Germany", "France"],
    "Age": [44, 27, 30, 38, 40, 35, None, 48, 50, 37],
    "Salary": [72000, 48000, 54000, 61000, None, 58000, 52000, 79000, 83000, 67000],
    "Purchased": ["No", "Yes", "No", "No", "Yes", "Yes", "No", "Yes", "No", "Yes"]
}

df = pd.DataFrame(data)

# Apply one-hot encoding on the 'Country' column
onehot_df = pd.get_dummies(df, columns=['Country'])

# Apply label encoding on the 'Purchased' column
le = LabelEncoder()
onehot_df['Purchased'] = le.fit_transform(onehot_df['Purchased'])

onehot_df
```

```python
import numpy as np

def weighted_average(arr, weights, axis):
    return np.average(arr, axis=axis, weights=weights)

# Example usage:
arr = np.array([1, 2, 3, 4, 5])
weights = np.array([0.1, 0.2, 0.3, 0.2, 0.2])
axis = 0

result = weighted_average(arr, weights, axis)
print(result)
```

```python
import pandas as pd

# Create a dictionary to store the students' information
data = {
    'Name': ['John', 'Emily', 'Michael', 'Sarah', 'David'],
    'Graduation Percentage': [85, 92, 78, 95, 88],
    'Age': [22, 20, 21, 23, 22]
}

# Create a Pandas DataFrame from the dictionary
df = pd.DataFrame(data)

# Calculate and display the average age of students
avg_age = df['Age'].mean()
print("Average Age of Students: " + str(avg_age))

# Calculate and display the average graduation percentage
avg_graduation_percentage = df['Graduation Percentage'].mean()
print("Average Graduation Percentage: " + str(avg_graduation_percentage))
```
